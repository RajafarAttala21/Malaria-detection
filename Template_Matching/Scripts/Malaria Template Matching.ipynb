Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.patches import Circle, Rectangle
from skimage import transform
from skimage.io import imread, imshow
from skimage.color import rgb2gray
from skimage.feature import match_template
from skimage.feature import peak_local_max
```

Load image then change to grayscale

```python
blood = imread(
    'D:\\Work\\Tech\\Data Science\\Python\\Notebook\\Malaria detection\\Template_Matching\\Images\\redbloodcells.png')
plt.figure(num=None, figsize=(14, 7))
plt.axis('off')
plt.title('Original', fontsize=14)
imshow(blood)
```

```python
blood_gray = rgb2gray(blood)
plt.figure(num=None, figsize=(14, 7))
plt.axis('off')
plt.title('Grayscale', fontsize=14)
imshow(blood_gray)
```

Cut the image to be used as a Template Image according to the coordinates

```python
template = blood_gray[155:200, 24:70]
plt.axis('off')
plt.title('Template Image', fontsize=14)
imshow(template)
```

Image grayscale map

```python
result_image = match_template(blood_gray, template)
plt.figure(num=None, figsize=(14, 7))
plt.axis('off')
plt.title('Map', fontsize=14)
imshow(result_image)
```

Make a red marker to match the cropped image

```python
x, y = np.unravel_index(np.argmax(result_image), result_image.shape)
template_width, template_height = template.shape
rect = plt.Rectangle((y, x), template_height,
                     template_width, color='r', fc='none')
plt.figure(num=None, figsize=(8, 6))
plt.gca().add_patch(rect)
plt.axis('off')
plt.title('Result', fontsize=14)
imshow(blood_gray)
```
