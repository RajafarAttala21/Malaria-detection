Import libraries

```python
import matplotlib.pyplot as plt
from skimage.filters import threshold_otsu, roberts
from skimage.io import imread
from skimage.color import rgb2gray
from skimage.feature import canny
```

Load image

```python
image = imread(
    'D:\\Work\\Tech\\Data Science\\Python\\Notebook\\Malaria detection\\Thresholding\\Images\\redbloodcells.PNG')
original = rgb2gray(image)
```

Binary Thresholding

```python
thresh = threshold_otsu(original)
binary = original > thresh
```

Roberts Thresholding

```python
newthresh = roberts(original)
```

Canny Thresholding

```python
newcanny = canny(original, sigma=1, high_threshold=0.6, low_threshold=0.6 ,use_quantiles=False)
```

Visualization

```python
fig, axes = plt.subplots(ncols=5, figsize=(25, 15))
ax = axes.ravel()

ax[0] = plt.subplot(2, 5, 1)
ax[1] = plt.subplot(2, 5, 2)
ax[2] = plt.subplot(2, 5, 3, sharex=ax[0], sharey=ax[0])
ax[3] = plt.subplot(2, 5, 4, sharex=ax[0], sharey=ax[0])
ax[4] = plt.subplot(2, 5, 5, sharex=ax[0], sharey=ax[0])

ax[0].imshow(original, cmap=plt.cm.gray)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].hist(original.ravel(), bins=256)
ax[1].set_title('Histogram')
ax[1].axvline(thresh, color='r')

ax[2].imshow(binary, cmap=plt.cm.gray)
ax[2].set_title('Thresholded Otsu (Binary)')
ax[2].axis('off')

ax[3].imshow(newthresh, cmap=plt.cm.gray)
ax[3].set_title('Thresholded Roberts')
ax[3].axis('off')

ax[4].imshow(newcanny, cmap=plt.cm.gray)
ax[4].set_title('Thresholded Canny')
ax[4].axis('off')

plt.show()
```
